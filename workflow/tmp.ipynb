```python
fin2 = "data/10_fig1-HYDG-tree/fam0023_tree.tsv"
df2 = pd.read_csv(fin2, sep="\t")
```

```python
# filter rows that Kingdom is NaN
df3 = df2[~df2.Kingdom.notnull()]
```

```python
# length(label) > 5
df4 = df3[df3.label.str.len() > 5]
```

```python
# label not start with "IMGVR"
df5 = df4[df4.label.str.startswith("IMGVR")]
```

```python
# split label by "|" and get the first element as new column "UVIG"
df5["UVIG"] = df5.label.str.split("|").str[0]
```

```python
df5.UVIG.nunique()
```

```python
# save df5 unique UVIG to file
fout = "data/10_fig1-HYDG-tree/IMGVR_HYDG_contigs.tsv"
df5.UVIG.unique().tofile(fout, sep="\n")
```

```python
# label not start with "NNKMIP_" or "JABKED_"
df6 = df5[~df5.label.str.startswith("NNKMIP_")]
df7 = df6[~df6.label.str.startswith("JABKED_")]
```
